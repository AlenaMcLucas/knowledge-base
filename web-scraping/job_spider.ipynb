## Apply to Whole Dataset

```python
job_des = jobs['Description']
```

```python
job_des = job_des.apply(lambda x: TAG_RE.sub(' ', x))
```

```python
job_des
```

```python
job_des[250]
```

```python
job_tokens = job_des.apply(lambda x: pos_tag(word_tokenize(x)))
```

```python
job_tokens
```

```python
def noun_count(pos_tokens):
    nouns = []

    for tok in pos_tokens:
        if ( tok[1] in ('NN', 'NNS', 'NNP', 'NNPS') ) and ( tok[0].lower() not in stopwords.words('english')):
            nouns.append(tok[0].lower())
            
    word_count = {i:nouns.count(i) for i in set(nouns)}
    
    return word_count
```

```python
job_count = job_tokens.apply(lambda x: noun_count(x))
```

```python
job_count
```

```python
def noun_set(pos_tokens):
    nouns = []

    for tok in pos_tokens:
        if ( tok[1] in ('NN', 'NNS', 'NNP', 'NNPS') ) and ( tok[0].lower() not in stopwords.words('english')):
            nouns.append(tok[0].lower())
            
    word_count = Counter({i:1 for i in set(nouns)})
    
    return word_count
```

```python
job_set = job_tokens.apply(lambda x: noun_set(x))
```

```python
job_set
```

```python
job_set_count = Counter({})

for i in np.arange(len(job_set)):
    job_set_count += job_set[i]
```
